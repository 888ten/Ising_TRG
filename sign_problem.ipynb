# 符号問題


符号問題を抱える積分の例として$\frac{\int dx \, x^2 e^{-\frac{1}{2}x^2-ipx}}{\int dx \,  e^{-\frac{1}{2}x^2-ipx}}=1-p^2$
をモンテカルロ法で計算します。また、その計算結果を$1-p^2$で割った値をプロットします。


そして虚部が大きくなるにつれ、計算結果が実際の積分の値では無くなっていくことを確認します。

```python
import numpy as np
import matplotlib.pyplot as plt

denominator = lambda x, p: np.exp(-1.0j*p*x)
numerator = lambda x, p: x**2 * np.exp(-1.0j*p*x)
error_ratio = lambda x, p: np.abs(numerator(x, p) / denominator(x, p) / (1 - p**2))

Nsample = 10000
list_p = np.arange(0, 50, 2)
list_obs = [np.mean([error_ratio(np.random.normal(0.0, 1.0, 1), p) for i in np.arange(Nsample)]) for p in list_p]
plt.plot(list_p, list_obs, color="purple")
plt.xlabel('p')
plt.ylabel('Accuracy of calculation')
plt.yscale('log')
plt.show()
```
